```python
num=int(input("enter the number:"))
def reverseee(num):
    actual=num
    rev=0
    while actual:
        digit=actual%10
        rev=rev*10+digit
        actual//=10
    return rev
print(reverseee(num))
```

```python
n,m=map(int,input("enter number").split())
def swappp(a,b):
    a=a^b
    b=a^b
    a=a^b
    return a,b
print(swappp(n,m))
```

```python
def roman(rom):
    d={
        "I":1,
        "V":5,
        "X":10,
        "L":50,
        "C":100,
        "D":500,
        "M":1000
    }
    total,curr,prev=0,0,0
    for i in reversed (rom):
        curr=d[i]
        if curr<prev:
            total-=curr
        else:
            total+=curr
        prev=curr
    return total
rom=input("enter the roman number")
print(roman(rom))
    
```

```python
arr=[2,3,4,5,12,9,6,10]
target=7

def twoSumm(arr,target):
    for i in range(len(arr)):
        diff=target-arr[i]
        if diff in arr:
            if diff==5:
                continue
            return [arr[i],diff]
           
print(twoSumm(arr,target))
        
    
    
    
```
