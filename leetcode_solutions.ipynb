### 189 Rotate Array

```python
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        n=len(nums)
        k=k%n
        self.reverse(nums,0,n-1)
        self.reverse(nums,0,k-1)
        self.reverse(nums,k,n-1)
    def reverse(self,nums,start,end):
        i=start
        j=end
        while(i<j):
            nums[i],nums[j]=nums[j],nums[i]
            i+=1
            j-=1
```

### Max Consecutive Ones

```python
class Solution:
    def findMaxConsecutiveOnes(self, nums: List[int]) -> int:
        ans=0
        count=0
        for i in nums:
            if i==1:
                count+=1
            else:
                count=0
            if count>ans:
                ans=count
        return ans
        
```

### Product of an Array

```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        n=len(nums)
        left=[1]*n
        right=[1]*n
        for i in range (1,n):
            left[i]=left[i-1]*nums[i-1]
        
        for i in range(n-2,-1,-1):
            right[i]=right[i+1]*nums[i+1]

        res=[]
        for i in range(0,n):
            res.append(left[i]*right[i])
        return res
```

### 53. Maximum Subarray

```python
class Solution:
    def maxsubarray(self,nums:List[int])->int:
        ans=nums[0]
        curr=nums[0]
        for i in range(1,len(nums)):
            if (curr<0):
                curr=0
            curr+=nums[i]
            if (curr>ans):
                ans=curr
        return ans
    
print(Solution().maxsubarray([-2,1,-3,4,-1,2,1,-5,4]))
```

### 152. Maximum Product Subarray

```python
from typing import List
class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        ans=nums[0]
        maxP=nums[0]
        minP=nums[0]
        for i in range(1,len(nums)):
            if (nums[i]>=0):
                maxP=max(maxP*nums[i],nums[i])
                minP=min(minP*nums[i],nums[i])
            else:
                temp=maxP
                maxP=max(minP*nums[i],nums[i])
                minP=min(temp*nums[i],nums[i])
            ans=max(ans,maxP)
        return ans
print(Solution().maxProduct([2,3,-2,4]))
```

### 36 Valid Sudoku

```python
class Solution:
    def isValidSudoku(self, board: List[List[str]]) -> bool:
        rowSet = [set() for _ in range(9)]
        colSet = [set() for _ in range(9)]
        gridSet = [set() for _ in range(9)]

        for i in range(9):
            for j in range(9):
                if board[i][j] == '.':
                    continue

                gridNo = (i // 3) * 3 + (j // 3)

                isPresentInRow = board[i][j] in rowSet[i]
                isPresentInCol = board[i][j] in colSet[j]
                isPresentInGrid = board[i][j] in gridSet[gridNo]

                if (isPresentInRow or isPresentInCol or isPresentInGrid):
                    return False

                rowSet[i].add(board[i][j])
                colSet[j].add(board[i][j])
                gridSet[gridNo].add(board[i][j])

        return True
```

### 283 Move Zeros

```python
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        start=0
        for i in range(len(nums)):
            if nums[i]!=0:
                nums[i],nums[start]=nums[start],nums[i]
                start+=1
```
