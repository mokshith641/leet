### 189 Rotate Array

```python
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        n=len(nums)
        k=k%n
        self.reverse(nums,0,n-1)
        self.reverse(nums,0,k-1)
        self.reverse(nums,k,n-1)
    def reverse(self,nums,start,end):
        i=start
        j=end
        while(i<j):
            nums[i],nums[j]=nums[j],nums[i]
            i+=1
            j-=1
```

### Product of an Array

```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        n=len(nums)
        left=[1]*n
        right=[1]*n
        for i in range (1,n):
            left[i]=left[i-1]*nums[i-1]
        
        for i in range(n-2,-1,-1):
            right[i]=right[i+1]*nums[i+1]

        res=[]
        for i in range(0,n):
            res.append(left[i]*right[i])
        return res
```

### 53. Maximum Subarray

```python
class Solution:
    def maxsubarray(self,nums:List[int])->int:
        ans=nums[0]
        curr=nums[0]
        for i in range(1,len(nums)):
            if (curr<0):
                curr=0
            curr+=nums[i]
            if (curr>ans):
                ans=curr
        return ans
    
print(Solution().maxsubarray([-2,1,-3,4,-1,2,1,-5,4]))
```

### 152. Maximum Product Subarray

```python
from typing import List
class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        ans=nums[0]
        maxP=nums[0]
        minP=nums[0]
        for i in range(1,len(nums)):
            if (nums[i]>=0):
                maxP=max(maxP*nums[i],nums[i])
                minP=min(minP*nums[i],nums[i])
            else:
                temp=maxP
                maxP=max(minP*nums[i],nums[i])
                minP=min(temp*nums[i],nums[i])
            ans=max(ans,maxP)
        return ans
print(Solution().maxProduct([2,3,-2,4]))
```

### 36 Valid Sudoku

```python
class Solution:
    def isValidSudoku(self, board: List[List[str]]) -> bool:
        rowSet = [set() for _ in range(9)]
        colSet = [set() for _ in range(9)]
        gridSet = [set() for _ in range(9)]

        for i in range(9):
            for j in range(9):
                if board[i][j] == '.':
                    continue

                gridNo = (i // 3) * 3 + (j // 3)

                isPresentInRow = board[i][j] in rowSet[i]
                isPresentInCol = board[i][j] in colSet[j]
                isPresentInGrid = board[i][j] in gridSet[gridNo]

                if (isPresentInRow or isPresentInCol or isPresentInGrid):
                    return False

                rowSet[i].add(board[i][j])
                colSet[j].add(board[i][j])
                gridSet[gridNo].add(board[i][j])

        return True
```

### 169 Majority Element

```python
class Solution:
    def majorityElement(self, nums):
        me=nums[0]
        count=1
        for i in range(1,len(nums)):
            if nums[i]==me:
                count+=1
            else:
                count-=1
                if count==0:
                    me=nums[i]
                    count=1
        return me
        
```

### 75 Sort Colors

```python
class Solution:
    def sortColors(self, nums):
        """
        Do not return anything, modify nums in-place instead.
        """
        left=0
        right=len(nums)-1
        i=0
        while i<=right:
            if nums[i]==0:
                nums[i],nums[left]=nums[left],nums[i]
                i+=1
                left+=1
            elif nums[i]==2:
                nums[right],nums[i]=nums[i],nums[right]
                right-=1
                
            
            else:
                i+=1
```

### 160 Intersection of Two Linked Lists

```python
class Solution:
    def getIntersectionNode(self, headA, headB):
        def findLen(self,head):
            length=0
            curr=head
            while(curr is not None):
                curr=curr.next
                length+=1
            return length
        m=findLen(self,headA)
        n=findLen(self,headB)
        fp=headA
        sp=headB
        if m<=n :
            for i in range(n-m):
                sp=sp.next
        else:
            for i in range(m-n):
                fp=fp.next
        while (sp is not None):
            if (fp==sp):
                return fp
            fp=fp.next
            sp=sp.next
        return None
        
```

### 141 Linked List Cycle

```python
class Solution:
    def hasCycle(self, head):
        if (head is None or head.next is None):
            return False
        first=head
        second=head.next.next
        while first and second and second.next:
            if first==second:
                return True
            first=first.next
            second=second.next.next
        return False
        
```

### 206 Reverse a Linked List (iteration and recursion)

```python
class Solution:
    def reverseList(self, head):
        prev=None
        curr=head
        while curr:
            next_node=curr.next
            curr.next=prev
            prev=curr
            curr=next_node
        return prev
```

### 234 Palindrome Linked List

```python
def isPalindrome(self, head):
        slow=head
        fast=head
        while fast and fast.next:
            slow=slow.next
            fast=fast.next.next
        
        prev=None
        curr=slow

        while curr:
            temp=curr.next
            curr.next=prev
            prev=curr
            curr=temp
        
        left=head
        right=prev

        while right:
            if left.val!=right.val:
                return False
            left=left.next
            right=right.next
        return True
        
```

### 9 Palindrome Number

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        if x<0:
            return False
        original=x
        rev=0
        while original:
            digit=original%10
            rev=rev*10+digit
            original//=10
        if x==rev:
            return True
        return False
        
```

```python
def detectCycle(self, head):
        slow=head
        fast=head
        insPoint=None

        while True:
            if fast is None or fast.next is None:
                return None 
            fast=fast.next.next
            slow=slow.next
            if slow==fast:
                insPoint=slow
                break
        start=head
        while (start!=insPoint):
            start=start.next
            insPoint=insPoint.next
        return start
        
```

### 876 Middle of the linked list

```python
def middleNode(self, head):
        slow=head
        fast=head
        while fast and fast.next:
            fast=fast.next.next
            slow=slow.next
        return slow
```
